```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from scipy import stats
import plotly.express as px
from datetime import datetime, timedelta
```

# 1. Network Traffic (Simulated) EDA

```python
# Configuration
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams.update({'font.size': 12, 'figure.figsize': (14, 8)})
sns.set_palette("colorblind")
```

```python
# Load simulated data
df = pd.read_csv('network_traffic_simulation.csv')
```

```python
# Convert timestamp to datetime
base_date = datetime(2024, 1, 1)
df['timestamp'] = df['timestamp'].apply(lambda x: base_date + timedelta(minutes=x))
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

## 1.1 Dataset Overview

```python
print("="*80)
print("Dataset Overview")
print("="*80)
print(f"Time Range: {df['timestamp'].min()} to {df['timestamp'].max()}")
print(f"Total Records: {len(df):,}")
print(f"Time Resolution: {pd.Timedelta(minutes=5)}")
print("\nMissing Values:")
print(df.isnull().sum())
print("\nData Types:")
print(df.dtypes)
```

# 1.2 Temporal Analysis

```python

print("\n" + "="*80)
print("Temporal Analysis")
print("="*80)

```

**Create time-based features**

```python

df['hour'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.dayofweek
df['day_part'] = pd.cut(df['hour'], 
                        bins=[0, 6, 12, 18, 24],
                        labels=['Night', 'Morning', 'Afternoon', 'Evening'],
                        include_lowest=True)
```

**Academic phase distribution**

```python
phase_dist = df['phase'].value_counts(normalize=True)
print(f"\nAcademic Phase Distribution:\n{phase_dist}")
```

**Plot traffic patterns**

```python

fig, axes = plt.subplots(3, 1, figsize=(16, 18))
# Daily traffic pattern
daily_volume = df.groupby('hour')['total_volume'].mean()
sns.lineplot(x=daily_volume.index, y=daily_volume.values, ax=axes[0])
axes[0].set_title('Average Hourly Traffic Volume (MB/5-min)')
axes[0].set_ylabel('Volume (MB)')
axes[0].axvspan(10, 16, alpha=0.2, color='green', label='Peak Hours (10AM-4PM)')
axes[0].legend()
# Phase comparison
phase_volume = df.groupby(['phase', 'hour'])['total_volume'].mean().reset_index()
sns.lineplot(data=phase_volume, x='hour', y='total_volume', hue='phase', 
             ax=axes[1], linewidth=2.5)
axes[1].set_title('Traffic Volume by Academic Phase')
axes[1].set_ylabel('Volume (MB)')
# Weekly pattern
try:
    # Modern pivot syntax
    week_volume = df.groupby(['day_of_week', 'day_part'], observed=True)['total_volume'].mean().reset_index()
    pivot_table = week_volume.pivot(
        index='day_part', 
        columns='day_of_week', 
        values='total_volume'
    )
    
    # Create heatmap
    sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="YlGnBu", ax=axes[2])
    axes[2].set_title('Weekly Traffic Patterns (Volume in MB)')
    axes[2].set_yticklabels(['Night', 'Morning', 'Afternoon', 'Evening'])
    axes[2].set_xticklabels(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
except Exception as e:
    print(f"Error creating weekly pattern heatmap: {str(e)}")
    # Fallback visualization
    avg_daily = df.groupby('day_of_week')['total_volume'].mean()
    axes[2].bar(avg_daily.index, avg_daily.values)
    axes[2].set_title('Average Daily Volume')
    axes[2].set_xticks(range(7))
    axes[2].set_xticklabels(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
    axes[2].set_ylabel('Volume (MB)')

plt.tight_layout()
plt.savefig('temporal_analysis.png', dpi=300)
plt.show()
```

## 1.3. Time Series Decomposition

```python
print("\n" + "="*80)
print("Time Series Decomposition")
print("="*80)

# Select only numerical columns
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

if 'total_volume' in numeric_cols:
    # Resample to hourly data using correct frequency
    hourly_df = df.set_index('timestamp')[numeric_cols].resample('h').mean()
    
    if not hourly_df.empty:
        # Perform decomposition only on total_volume
        try:
            # Ensure we have enough data points
            if len(hourly_df) > 48:  # At least 2 days
                decomposition = seasonal_decompose(
                    hourly_df['total_volume'].dropna(), 
                    model='additive', 
                    period=24  # Daily seasonality
                )
                
                fig, axes = plt.subplots(4, 1, figsize=(14, 12))
                decomposition.observed.plot(ax=axes[0], title='Observed')
                decomposition.trend.plot(ax=axes[1], title='Trend')
                decomposition.seasonal.plot(ax=axes[2], title='Seasonality (24-hour)')
                decomposition.resid.plot(ax=axes[3], title='Residuals')
                plt.tight_layout()
                plt.savefig('time_series_decomposition.png', dpi=300)
                plt.show()
            else:
                print("Insufficient data for decomposition (need >48 hours)")
        except ValueError as e:
            print(f"Decomposition failed: {str(e)}")
        except Exception as e:
            print(f"Unexpected error: {str(e)}")
    else:
        print("Hourly dataframe is empty after resampling")
else:
    print("'total_volume' column not found in numeric columns")
```

## 1.4. Class-Specific Analysis

```python
print("\n" + "="*80)
print("Class-Specific Analysis")
print("="*80)
```


**Prepare class data**

```python
class_data = []
for class_name in ['Teaching', 'Non-teaching', 'Managerial', 'Non-managerial']:
    class_df = df[[f'{class_name}_requests', f'{class_name}_volume']].copy()
    class_df['class'] = class_name
    class_df.rename(columns={f'{class_name}_requests': 'requests', 
                           f'{class_name}_volume': 'volume'}, inplace=True)
    class_data.append(class_df)
    
class_df = pd.concat(class_data, ignore_index=True)
```

**Summary statistics**

```python
print("\nSummary Statistics by Class:")
print(class_df.groupby('class').describe().transpose())
```

**Distribution plots**

```python

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Volume distribution
sns.boxplot(data=class_df, x='class', y='volume', showfliers=False, ax=axes[0,0])
axes[0,0].set_title('Traffic Volume Distribution by Class')
axes[0,0].set_ylabel('Volume (MB)')

# Request distribution
sns.violinplot(data=class_df, x='class', y='requests', inner="quartile", 
               cut=0, ax=axes[0,1])
axes[0,1].set_title('Request Distribution by Class')
axes[0,1].set_ylabel('Requests')

# Phase comparison
phase_class_volume = []
for class_name in ['Teaching', 'Non-teaching', 'Managerial', 'Non-managerial']:
    for phase in ['Lecture', 'Examination', 'Administrative']:
        mean_vol = df[df['phase'] == phase][f'{class_name}_volume'].mean()
        phase_class_volume.append({'class': class_name, 'phase': phase, 'volume': mean_vol})
        
phase_class_volume = pd.DataFrame(phase_class_volume)
sns.barplot(data=phase_class_volume, x='class', y='volume', hue='phase', 
            ax=axes[1,0], errorbar=None)
axes[1,0].set_title('Mean Volume by Class and Academic Phase')
axes[1,0].set_ylabel('Volume (MB)')

# Volume over time (example: Teaching class)
sns.lineplot(data=df, x='timestamp', y='Teaching_volume', 
             ax=axes[1,1], alpha=0.7)
axes[1,1].set_title('Teaching Volume Over Time')
axes[1,1].set_ylabel('Volume (MB)')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('class_analysis.png', dpi=300)
plt.show()
```

## 1.5 Policy Weight and Utilization Analysis

```python
print("\n" + "="*80)
print("Policy Weight Analysis")
print("="*80)
```

**Teaching weight during examinations**

```python
exam_df = df[df['phase'] == 'Examination'].copy()
if not exam_df.empty:
    peak_time = exam_df.loc[exam_df['Teaching_weight'].idxmax(), 'timestamp']
    exam_df['days_to_peak'] = (exam_df['timestamp'] - peak_time).dt.days.abs()
    
    # Plot weight progression
    plt.figure(figsize=(12, 6))
    sns.regplot(data=exam_df, x='days_to_peak', y='Teaching_weight', 
                lowess=True, scatter_kws={'alpha':0.3})
    plt.title('Teaching Weight Proximity Effect During Examinations')
    plt.xlabel('Days from Peak Examination Date')
    plt.ylabel('Teaching Weight')
    plt.axhline(y=0.28, color='r', linestyle='--', label='Base Weight')
    plt.legend()
    plt.savefig('weight_progression.png', dpi=300)
    plt.show()
else:
    print("Warning: No examination phase data found")
```

**Utilization analysis**

```python
print(f"\nOverall Utilization: {df['utilization'].mean():.2%}")
print(f"Utilization by Capacity Period:")
print(df.groupby(df['timestamp'].dt.year)['utilization'].mean().apply(lambda x: f"{x:.2%}"))
```

## 1.6. Correlation and Feature Analysis

```python
print("\n" + "="*80)
print("Correlation Analysis")
print("="*80)
```

**Feature engineering for Q-learning**

```python
# 
df['Teaching_demand_ratio'] = df['Teaching_volume'] / df['capacity']
df['Managerial_demand_ratio'] = df['Managerial_volume'] / df['capacity']
df['phase_code'] = df['phase'].map({'Lecture':0, 'Examination':1, 'Administrative':2})
```

**Correlation matrix**

```python
corr_cols = ['total_volume', 'utilization', 'Teaching_volume', 'Non-teaching_volume',
             'Managerial_volume', 'Non-managerial_volume', 'Teaching_weight', 
             'Teaching_demand_ratio', 'phase_code']
corr_matrix = df[corr_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", center=0,
            mask=np.triu(np.ones_like(corr_matrix, dtype=bool)))
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.savefig('correlation_matrix.png', dpi=300)
plt.show()
```

## 1.7. Statistical Distribution Validation

```python
print("\n" + "="*80)
print("Statistical Distribution Validation")
print("="*80)
```

**Teaching session duration (power-law)**

```python
teaching_volume = df['Teaching_volume'].values
teaching_volume = teaching_volume[teaching_volume > 0]  # Remove zeros

if len(teaching_volume) > 0:
    try:
        # Power-law fit
        fit_params = stats.powerlaw.fit(teaching_volume, floc=0)
        print(f"Teaching Volume Power-law Fit: alpha={fit_params[0]:.2f} (Theoretical: 2.3)")
        print(f"KS Test vs Power-law: {stats.kstest(teaching_volume, 'powerlaw', args=fit_params)}")
    except Exception as e:
        print(f"Power-law fitting error: {str(e)}")
else:
    print("No Teaching volume data available for distribution fitting")
```

**Generate comprehensive EDA report**

```python
report_date = datetime.now().strftime('%Y%m%d_%H%M%S')
report_filename = f"traffic_simulation_eda_report_{report_date}.txt"
# Use UTF-8 encoding to handle special characters
with open(report_filename, 'w', encoding='utf-8') as f:
    f.write("Network Traffic Simulation EDA Report\n")
    f.write("="*50 + "\n\n")
    f.write(f"Generated on: {datetime.now()}\n\n")
    
    
    if len(teaching_volume) > 0:
        f.write("\n\n5. Distribution Validation\n")
        f.write("-"*50 + "\n")
        # Use alpha instead of Î± character
        f.write(f"Teaching Volume Power-law Fit: alpha={fit_params[0]:.2f} (Theoretical: 2.3)\n")
        f.write(f"KS Test vs Power-law: {stats.kstest(teaching_volume, 'powerlaw', args=fit_params)}\n")
    
    # Add cross-correlation summary
    f.write("\n\n6. Cross-correlation Insights\n")
    f.write("-"*50 + "\n")
    f.write("Peak cross-correlation between class traffic and total volume:\n")
    for class_name in ['Teaching', 'Managerial']:
        ccf = np.correlate(df[f'{class_name}_volume'] - df[f'{class_name}_volume'].mean(),
                           df['total_volume'] - df['total_volume'].mean(),
                           mode='full')
        max_corr = np.max(ccf)
        f.write(f"- {class_name}: {max_corr:.3f} at lag {np.argmax(ccf) - len(ccf)//2}\n")

print(f"\nEDA Complete. Visualizations and report saved as '{report_filename}'.")
```
